# 1. Download the Datasets

This notebook downloads the benchmark datasets used in the project and stores them
in `data/raw/`. All datasets come from the Hugging Face Hub with their **official
train/validation/test splits**, so results stay comparable with published work.

| Dataset | Task | Platform |
|---------|------|----------|
| TweetEval (sentiment) | 3-class sentiment | Twitter |
| GoEmotions | 28-label emotion (multi-label) | Reddit |

```python
import os
from datasets import load_dataset

RAW = "../data/raw"
os.makedirs(RAW, exist_ok=True)

# name we use locally  ->  (repository on Hugging Face, configuration)
DATASETS = {
    "tweeteval_sentiment": ("cardiffnlp/tweet_eval", "sentiment"),
    "goemotions":          ("google-research-datasets/go_emotions", "simplified"),
}
```

```python
# download each dataset (skips automatically if already saved)
for name, args in DATASETS.items():
    path = os.path.join(RAW, name)
    if os.path.exists(path):
        print(f"[skip] {name} is already in {path}")
        continue
    print(f"[downloading] {name} ...")
    ds = load_dataset(*args)
    ds.save_to_disk(path)
    for split, data in ds.items():
        print(f"   {split:<12} {len(data):>7} examples")
print("\nDone - all raw data is in data/raw/")
```

```python
# quick check: load everything back and look at one example of each
from datasets import load_from_disk

for name in DATASETS:
    ds = load_from_disk(os.path.join(RAW, name))
    sizes = {k: len(v) for k, v in ds.items()}
    print(f"{name:<22} {sizes}")
    print("   example:", str(ds["train"][0])[:100], "...\n")
```
